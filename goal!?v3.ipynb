```python
import pandas as pd
import numpy as np
import random
```

```python
#Dims
X = 15
Y = 15
```

```python
start = []
for i in range(0,X):
    across = []
    for j in range(0,Y):
        input = random.randint(0,1)
        if input == 1:
           input = "A"
        else:
           input = "B"
        across.append(input)
    start.append(across)

start = pd.DataFrame(start)
```

##Starting Board
 - Dimensions are X * Y

```python
def check(input,dim):
    if input < 0:
       input = 14
    if input > dim - 1:
       input = 0
    return input

height = start.shape[1]
width = start.shape[1]
bb = {'O':[],'D1':[],'D2':[]}
bc = []
at = []
bt = []
ct = []
dt = []
et = []
ft = []
gt = []
ht = []
abo = []
bbo = []
cbo = []
dbo = []
ebo = []
fbo = []
gbo = []
hbo = []
for row in range(0, height):
    for col in range(0, width):
        
        apos = [check(row+1,height),check(col-1,width)]
        bpos = [check(row+1,height),check(col,width)]
        cpos = [check(row+1,height),check(col+1,width)]
        dpos = [check(row,height),check(col+1,width)]
        hpos = [check(row,height),check(col-1,width)]
        gpos = [check(row-1,height),check(col-1,width)]
        fpos = [check(row-1,height),check(col,width)]
        epos = [check(row-1,height),check(col+1,width)]
        batch = [apos,bpos,cpos,dpos,hpos,gpos,fpos,epos]
        at.append(apos)
        bt.append(bpos)
        ct.append(cpos)
        dt.append(dpos)
        et.append(epos)
        ft.append(fpos)
        gt.append(gpos)
        ht.append(hpos)
        bout = ["aval","bval","cval","dval","hval","gval","fval","eval"]
        j = 0
        foc = [row,col]
        bb['O'].append(foc)
        bb['D1'].append(batch)
        for i in batch:
            x = i[0]
            y = i[1]
            bout[j] = start.iloc[x][y]
            j += 1
        abo.append(bout[0])
        bbo.append(bout[1])
        cbo.append(bout[2])
        dbo.append(bout[3])
        ebo.append(bout[4])
        fbo.append(bout[5])
        gbo.append(bout[6])
        hbo.append(bout[7])
        bc.append(bout)
        bb['D2'].append(bout)

data = {'A':at,'B':bt,'C':ct,'D':dt,'E':et,'F':ft,'G':gt,'H':ht}
data2 = {'Av':abo, 'Bv':bbo, 'Cv':cbo, 'Dv':dbo, 'Ev':ebo, 'Fv':fbo, 'Gv':gbo, 'Hv':hbo}
be = pd.DataFrame(data)
bd = pd.DataFrame(bb)
bf = pd.DataFrame(data2)
```

```python
t = bb['D1']
ca = be.merge(bd,how='left',left_on=be.index,right_on=bd.index)
ca.drop(columns=['D1','key_0'], inplace=True)
ct = ca.copy()

cu = ct.merge(bf,how='left',left_on=ct.index,right_on=bf.index)
cu.drop(columns=['D2','key_0'],inplace=True)

cu = cu.reindex(sorted(cu.columns), axis=1)
cu
```
